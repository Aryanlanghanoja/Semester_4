```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
Dataset = pd.read_excel('./Dataset.xlsx','Python_Data')
```

```python
regression = linear_model.LinearRegression()
regression.fit(Dataset[["Hight"]], Dataset["Shoe Size(US)"])

print(f"B0 = {regression.intercept_} \nB1 = {regression.coef_}")
```

```python
plt.figure(figsize=(18,9))
plt.scatter(x=Dataset['Hight'], y=Dataset['Shoe Size(US)'], color='red', marker='*')
plt.plot(Dataset['Hight'], regression.predict(Dataset[["Hight"]]), color='blue')
plt.xlabel('Hight')
plt.ylabel('Shoe Size')
plt.title('Shoe Size Prediction')
plt.show()
```

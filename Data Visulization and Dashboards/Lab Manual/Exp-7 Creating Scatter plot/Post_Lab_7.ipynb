```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

Dataset = pd.read_excel(
    "D:/Aryan Data/Usefull Data/Semester - 4/Data Visulization and Dashboards/Lab Manual/Exp-7 Creating Scatter plot/Sample - Superstore.xlsx"
,'Orders')

print(Dataset)
```

```python
# Task 1: Create a scatterplot to compare sales to profit in the Superstore dataset:

plt.figure(figsize=(16,8))
sns.scatterplot(data=Dataset, x=Dataset['Profit'] , y=Dataset['Sales'], hue=Dataset['Segment'])
plt.title("Scatter Plot of Superstore Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.legend()
plt.show()
```

```python
# Task 2: Create a scatterplot to compare sales to quantity in the Superstore dataset:

plt.figure(figsize=(16, 8))
sns.scatterplot(x=Dataset["Quantity"], y=Dataset["Sales"], data=Dataset, hue=Dataset["Segment"])
plt.title("Task-2 Scatter Plot of Sales vs Quantity in the Superstore")
plt.legend()
plt.show()
```

```python
# Task 3: Create a scatterplot to compare Quantity to discount in the Superstore dataset:

plt.figure(figsize=(16, 8))
sns.scatterplot(x=Dataset["Quantity"], y=Dataset["Discount"], data=Dataset, hue=Dataset["Segment"])
plt.title("Task-3 Scatter Plot of Quantity vs Discount in the Superstore")
plt.legend()
plt.show()
```

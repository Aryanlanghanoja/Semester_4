```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

Dataset = pd.read_excel(
    "D:/Aryan Data/Usefull Data/Semester - 4/Data Visulization and Dashboards/Lab/Lab-5/Sample - Superstore.xlsx",
    "Orders",
)

Dataset["Order Date"] = pd.to_datetime(Dataset["Order Date"])
Dataset["Year"] = Dataset["Order Date"].dt.year
Dataset["Quater"] = Dataset["Order Date"].dt.quarter
Dataset["Month"] = Dataset["Order Date"].dt.month
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="Year", y="Sales", data=Dataset, estimator=sum)
plt.title("Order Date vs Sales(Years)")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="Year", y="Sales", data=Dataset, estimator=sum ,hue='Region')
plt.title("Order Date vs Sales(Years)")
plt.show()
```

```python
New_Dataset = Dataset.groupby(["Year", "Region"])["Sales"].sum().reset_index()
Sales_Central = []
Sales_South = []
Sales_East = []
Sales_West = []

for i in range(0, 16, 4):
    Sales_Central += [New_Dataset.iloc[i, 2]]

for i in range(2, 16, 4):
    Sales_South += [New_Dataset.iloc[i, 2]]

for i in range(1, 16, 4):
    Sales_East += [New_Dataset.iloc[i, 2]]

for i in range(3, 16, 4):
    Sales_West += [New_Dataset.iloc[i, 2]]

Year = [2014, 2015, 2016, 2017]

Sales_Central = np.array(Sales_Central)
Sales_East = np.array(Sales_East)
Sales_South = np.array(Sales_South)
Sales_West = np.array(Sales_West)

plt.bar(Year, Sales_Central, color="r")
plt.bar(Year, Sales_South, color="b",bottom=Sales_Central)
plt.bar(Year, Sales_East, color="g" , bottom=Sales_Central + Sales_South)
plt.bar(Year, Sales_West, color="y",bottom=Sales_Central + Sales_South + Sales_East)
plt.legend(['Central' , 'South','East','West'])
plt.show()


# plt.figure(figsize=(12, 6))
# sns.barplot(x="Year", y="Sales", data=Dataset, estimator=sum, hue="Region")
# plt.title("Order Date vs Sales(Years)")
# plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

Dataset = pd.read_excel(
    "D:/Aryan Data/Usefull Data/Semester - 4/Data Visulization and Dashboards/Lab/Lab-5/Sample - Superstore.xlsx",
    "Orders",
)

Dataset["Order Date"] = pd.to_datetime(Dataset["Order Date"])
Dataset["Year"] = Dataset["Order Date"].dt.year
Dataset["Quater"] = Dataset["Order Date"].dt.quarter
Dataset["Month"] = Dataset["Order Date"].dt.month
```

```python
# Task 1: Create a scatterplot to compare sales to profit in the Superstore dataset:

plt.figure(figsize = (16,8))
sns.scatterplot(x=Dataset["Profit"], y=Dataset["Sales"], data=Dataset  ,hue=Dataset['Segment'])
plt.title("Task-1 Scatter Plot of Sales vs Profit in the Superstore") 
plt.legend()
plt.show()
```

```python
# Task 2: Create a scatterplot to compare sales to quantity in the Superstore dataset:

plt.figure(figsize=(16, 8))
sns.scatterplot(
    x=Dataset["Quantity"], y=Dataset["Sales"], data=Dataset, hue=Dataset["Segment"]
)
plt.title("Task-2 Scatter Plot of Sales vs Quantity in the Superstore")
plt.legend()
plt.show()
```

```python
# Task 3: Create a scatterplot to compare Quantity to discount in the Superstore dataset:

plt.figure(figsize=(16, 8))
sns.scatterplot(x=Dataset["Quantity"], y=Dataset["Discount"], data=Dataset, hue=Dataset["Segment"])
plt.title("Task-3 Scatter Plot of Quantity vs Discount in the Superstore")
plt.legend()
plt.show()
```

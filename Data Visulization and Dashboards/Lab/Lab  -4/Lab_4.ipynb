```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

Dataset = pd.read_excel(
    "D:/Aryan Data/Usefull Data/Semester - 4/Data Visulization and Dashboards/Lab/Lab  -4/Sample - Superstore.xlsx",
    "Orders",
)

Dataset["Order Date"] = pd.to_datetime(Dataset["Order Date"])
Dataset["Year"] = Dataset["Order Date"].dt.year
Dataset["Quater"] = Dataset["Order Date"].dt.quarter
Dataset["Month"] = Dataset["Order Date"].dt.month
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x="Year", y="Sales", data=Dataset, estimator=sum)
plt.title("Order Date vs Sales(Years)")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x="Month", y="Sales", data=Dataset, estimator=sum)
plt.title("Order Date vs Sales(Years)")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x="Quater", y="Sales", data=Dataset, estimator=sum)
plt.title("Order Date vs Sales(Years)")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x = 'Quantity' , y = 'Profit' , data=Dataset , hue='State')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category',y = 'Quantity',data=Dataset ,hue='Sub-Category')
plt.show()
```
